```python
import numpy as np
import matplotlib.pyplot as plt
days = list(range(1, 11))
stock_price = [100, 102, 105, 103, 108, 112, 115, 117, 120, 125]

plt.plot(days, stock_price, marker="o", color="purple")
plt.xlabel("Day")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Over 10 Days")
plt.grid(True)
plt.show()
```
